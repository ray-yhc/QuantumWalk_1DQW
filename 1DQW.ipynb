```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
psi_0_initial = np.array([1,0]) # 초기값 설정. (psi_r , psi_l)
psi = np.zeros([2,201]) # position : -100 ~ +100 까지
psi[:,100] = psi_0_initial

hadamard = 1/np.sqrt(2) *np.array([[1,1],[1,-1]])

def C_hat(operator, psi_):
      #하나씩 따로 계산하는 방법
    #flipped = np.array(list(map(lambda x : operator@x ,psi_.T))).T
      #통으로 계산하는 방법
    flipped = (operator@psi_)
    # 둘 모두 동일한 결과가 나온다
    return flipped

def S_hat(psi_):
    shift_r = np.insert(psi_[0,:-1],0,0)
    shift_l = np.insert(psi_[1,1:],-1,0)
    return np.vstack((shift_r,shift_l))

def U_hat(operator, psi_):
    return S_hat(C_hat(operator, psi_))

def p_distribution(psi):
    dist = []
    for i in psi.T:
        p_n = i[0]*np.conj(i[0])+i[1]*np.conj(i[1])
        dist.append(p_n)
    return dist
```

```python
def Q_walker (psi_,walks, operator):
    results = []
    for i in range(0,walks):
        results.append(p_distribution(psi_)[1:-1])
        psi_ = U_hat(operator,psi_)
    return results
```

```python
result = Q_walker(psi,100,hadamard)
plt.figure(figsize=(15,8))
plt.subplot(221)
plt.bar(range(0,199),result[0])

plt.subplot(222)
plt.bar(range(0,199),result[5])

plt.subplot(223)
plt.bar(range(0,199),result[50])

plt.subplot(224)
plt.bar(range(0,199),result[99])
```

```python
def psi_i(right,left):
    init = np.array([right,left])
    result = np.zeros([2,201])
    result[:,100] = init
    return result
```

```python
result = Q_walker(psi_i(1,0),100,hadamard)
plt.figure(figsize=(15,8))
plt.subplot(221)
plt.bar(range(0,199),result[0])

plt.subplot(222)
plt.bar(range(0,199),result[5])

plt.subplot(223)
plt.bar(range(0,199),result[50])

plt.subplot(224)
plt.bar(range(0,199),result[99])
```
